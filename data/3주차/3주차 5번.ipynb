```python
!pip install koreanize_matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
# 데이터 불러오기
df = pd.read_csv("https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/NHIS_OPEN_GJ_2017.CSV.zip", encoding="cp949")
# 기준시력 9.9
df = df[(df['시력(좌)'] != 9.9) & (df['시력(우)'] != 9.9)]
# 연령대, 성별코드로나누고 시력좌우
df_grouped = df.groupby(['연령대코드(5세단위)', '성별코드'])[["시력(좌)", "시력(우)"]].mean().reset_index()
plt.figure(figsize=(14, 6))
# 왼눈 시력 시각화
plt.subplot(1, 2, 1)
sns.barplot(x='연령대코드(5세단위)', y='시력(좌)', hue='성별코드', data=df_grouped, palette={1: 'skyblue', 2: 'lightcoral'})
plt.title('연령대별 성별 좌측 평균 시력')
plt.ylim(0.1, 2.5)
# 오른눈 시력 시각화
plt.subplot(1, 2, 2)
sns.barplot(x='연령대코드(5세단위)', y='시력(우)', hue='성별코드', data=df_grouped, palette={1: 'skyblue', 2: 'lightcoral'})
plt.title('연령대별 성별 우측 평균 시력')
plt.ylim(0.1, 2.5)
plt.tight_layout()
plt.show()
```
